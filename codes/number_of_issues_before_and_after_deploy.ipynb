```python
def count_issues_month(year, list_dates):

    month_data = {month: [] for month in range(1, 13)}

    for date_str in list_dates:

          date = datetime.datetime.strptime(date_str, "%Y-%m-%d")

          month_data[date.month].append(date_str)



    issues_per_month = {datetime.date(year, month, 1).strftime('%B').lower(): len(issues) for month, issues in month_data.items()}

    return issues_per_month
```

```python
def before_and_after_dates(before_or_after_dates):

    issues_years = {}
    issues_years_months = {}
    issues_months = {}


    for dt in before_or_after_dates:
        date = datetime.datetime.strptime(dt, "%Y-%m-%d")

        if(date.year in issues_years):
            issues_years[date.year].append(dt)
        else:
            issues_years[date.year] = [dt]

    for year, dates in issues_years.items():
        data_issues = count_issues_month(year, dates)
        issues_years_months[year] = data_issues
#     print(commits_years_months)
    return issues_years_months

```
